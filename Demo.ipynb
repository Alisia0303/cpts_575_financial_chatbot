```python
import requests

headers = {"User-Agent": "YOUR EMAIL"}

def get_filing_links(cik, form_type="10-K", year=None):
    url = f"https://data.sec.gov/submissions/CIK{cik:010d}.json"
    data = requests.get(url, headers=headers).json()
    filings = data["filings"]["recent"]

    results = []
    for form, acc, doc, date in zip(filings["form"], filings["accessionNumber"],
                                    filings["primaryDocument"], filings["filingDate"]):
        if form == form_type and (year is None or date.startswith(str(year))):
            acc_no = acc.replace("-", "")
            link = f"https://www.sec.gov/Archives/edgar/data/{int(cik)}/{acc_no}/{doc}"
            results.append((date, link))
    return results

for idx, cik in enumerate(cik_500_dict.items()):
  try:
    filing_10K_url = get_filing_links(cik[1]['cik_str'], "10-K", 2024)[0][1]
    print("10-K filings:", cik[1]['ticker'], filing_10K_url)
  except:
    print("Fail to retrieve ", cik[1]['ticker'])
    continue
  # download 10-K filing as PDF
  pdf_10K_filing = pdfGeneratorApi.get_pdf(filing_10K_url)
  file_name = filing_10K_url.split('/')[-1].split('.')[0]
  # save PDF of 10-K filing to disk
  with open(f"/content/gdrive/My Drive/PHD/Courses/DS/SEC/{file_name}_10K.pdf", "wb") as file:
      file.write(pdf_10K_filing)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_text_splitters import CharacterTextSplitter

path = "/content/gdrive/My Drive/PHD/Courses/DS/SEC"
loader = PyPDFDirectoryLoader(path)
documents = loader.load()

# split it into chunks
text_splitter = CharacterTextSplitter(chunk_size=7000,
                                     chunk_overlap=700,
                                     separator='\n')
source_docs = text_splitter.split_documents(documents)
```

```python
#Add company and year metadata based on the filename of the source
for d in source_docs:
  metainfo = d.metadata['source'].split('/')[-1].split('_')
  d.metadata['company'] = metainfo[0].split('-')[0]
  d.metadata['year'] = int(metainfo[0].split('-')[1][:4])
  if len(metainfo) >= 4:
    d.metadata['Q-K'] = metainfo[1].split('.')[0]
  else:
    d.metadata['Q-K'] = "Annual"
```

```python
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
model_name = "BAAI/bge-small-en"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True} # set True to compute cosine simi
hf = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
db = Chroma.from_documents(source_docs, embedding=hf, persist_directory="/content/gdrive/My Drive/PHD/Courses/DS/db1")
```

```python
def docs_to_dict(docs,query,company):
  d={"question":query,
    "Company":company,
  }
  for i,(doc,score) in enumerate(docs):
    #d[f"page{i}_year"] = doc.metadata['year']
    d[f"page{i}"] = doc.page_content
    d[f"page{i}_Q-K"] = doc.metadata['Q-K']
  return d
```

```python
import pandas as pd
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import faithfulness, answer_correctness, context_precision, context_recall
from tqdm import tqdm
import random

retriever = db.as_retriever(search_kwargs={"k": 6})

# RAG Chain with retrieved chunks
prompt = PromptTemplate.from_template("""
You are a financial analyst. Answer only using the provided context.

Context:
{context}

Question: {question}

Answer with supporting chunks clearly cited.
""")

def format_docs(docs):
    return "\n\n".join([f"[Chunk {i+1} | {doc.metadata.get('company','?')} {doc.metadata.get('year','?')} {doc.metadata.get('Q-K','?')}]\n{doc.page_content}"
                       for i, doc in enumerate(docs)])

rag_chain = (
    {"context": retriever | format_docs, "question": lambda x: x}
    | prompt
    | llm
    | StrOutputParser()
)

# Load questions
df = pd.read_csv("financial_eval_5.csv")

# Run all systems
results = []

for _, row in tqdm(df.iterrows(), total=len(df), desc="Running evaluation"):
    q = row["question"]
    company = row["company"].lower()

    # Filter by company
    docs = db.similarity_search(q, k=6, filter={"company": company})
    context_str = format_docs(docs)

    # Your RAG
    rag_answer = rag_chain.invoke(q)

    # Baselines
    llm_only = llm.invoke(q).content
    bm25_docs = db.similarity_search(q, k=6)  # no filter
    bm25_answer = llm.invoke(f"Context:\n{format_docs(bm25_docs)}\nQuestion: {q}").content
    random_docs = random.sample(db._collection.get(include=["documents"])["documents"], 6)
    random_answer = llm.invoke(f"Context:\n{' '.join(random_docs[:3])}\nQuestion: {q}").content

    results.append({
        "question": q,
        "ground_truth": row["ground_truth_answer"],
        "rag_answer": rag_answer,
        "rag_contexts": [d.page_content for d in docs],
        "llm_only": llm_only,
        "bm25_answer": bm25_answer,
        "random_answer": random_answer
    })

# RAGAS Evaluation
ragas_ds = Dataset.from_list([{
    "question": r["question"],
    "answer": r["rag_answer"],
    "contexts": r["rag_contexts"],
    "ground_truth": r["ground_truth"]
} for r in results])

scores = evaluate(
    ragas_ds,
    metrics=[faithfulness, answer_correctness, context_precision, context_recall],
    llm=llm,
    embeddings=hf
)

print("\nRAG SCORES (50 questions):")
print(scores) # RAG model {'faithfulness': 1.0000, 'answer_correctness': 0.8085, 'context_precision': 0.9827, 'context_recall': 1.0000}
# Random: {'faithfulness': 0.0606, 'answer_correctness': nan, 'context_precision': 0.8550, 'context_recall': 1.0000}
# LLM: {'faithfulness': 0.3472, 'answer_correctness': 0.7195, 'context_precision': nan, 'context_recall': 1.0000}
# BM25 {'faithfulness': 0.8800, 'answer_correctness': 0.5199, 'context_precision': 1.0000, 'context_recall': 1.0000}


```
